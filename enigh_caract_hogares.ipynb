```
#Se importa la tabla "Características de las viviendas que habitan..." para su análsis.

caract_hogares = pd.read_csv('/content/conjunto_de_datos_viviendas_enigh2022_ns.csv', encoding='utf-8-sig')
caract_hogares.reset_index(drop=True, inplace=True)
caract_hogares.sample(10)
```

```
caract_hogares.info()
```

##Limpieza y modelado de datos

```
#Se estandarizan los datos contenidos en la columna "folioviv".

caract_hogares['folioviv'] = caract_hogares['folioviv'].astype(str)
caract_hogares['folioviv'] = caract_hogares['folioviv'].apply(
    lambda x: '0' + x if len(x) == 9 else x
)
caract_hogares['folioviv'].str.len().value_counts()
```

```
caract_hogares.sample(10)
```

```
#Creación de diccionario para crear nueva columna "entidad"
#Creación de columna "entidad"

entidades_mexico = {
    "01": "Aguascalientes",
    "02": "Baja California",
    "03": "Baja California Sur",
    "04": "Campeche",
    "05": "Coahuila",
    "06": "Colima",
    "07": "Chiapas",
    "08": "Chihuahua",
    "09": "Ciudad de México",
    "10": "Durango",
    "11": "Guanajuato",
    "12": "Guerrero",
    "13": "Hidalgo",
    "14": "Jalisco",
    "15": "México",
    "16": "Michoacán",
    "17": "Morelos",
    "18": "Nayarit",
    "19": "Nuevo León",
    "20": "Oaxaca",
    "21": "Puebla",
    "22": "Querétaro",
    "23": "Quintana Roo",
    "24": "San Luis Potosí",
    "25": "Sinaloa",
    "26": "Sonora",
    "27": "Tabasco",
    "28": "Tamaulipas",
    "29": "Tlaxcala",
    "30": "Veracruz",
    "31": "Yucatán",
    "32": "Zacatecas"
}


caract_hogares['entidad'] = caract_hogares['folioviv'].str[:2].map(entidades_mexico)
caract_hogares.sample(10)
```

```
#Se eliminan columnas con datos nulos o no incluidas en el análisis.

columnas_a_eliminar = caract_hogares.columns[[0, 6, 8, 9, 10, 12, 17, 18, 19, 22, 23, 28, 29, 30, 31, 35, 36, 37, 38, 53, 54, 55, 56, 59, 60, 61, 62, 63]]
caract_hogares = caract_hogares.drop(columns=columnas_a_eliminar)
caract_hogares.head()
```
