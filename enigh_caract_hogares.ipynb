##Análisis de proporciones

###Proporción de hogares con alguna carencia (agua, drenaje, electricidad)

```
caract_hogares['carencia_servicios'] = (
    (caract_hogares['lugar_agua'] == 3) |  # Agua fuera de la vivienda
    (caract_hogares['drenaje'] == 5) |     # No tiene drenaje
    (caract_hogares['disp_electricidad'] == 5)  # No tiene electricidad
)
```

```
caract_hogares.sample(5)
```

```
# Agrupar por entidad y calcular proporción de hogares con carencia
# Ordenar entidades de mayor a menor proporción de carencia
import plotly.io as pio

carencia_por_entidad = caract_hogares.groupby('entidad')['carencia_servicios'].mean().reset_index()
carencia_por_entidad = carencia_por_entidad.sort_values(by='carencia_servicios', ascending=False)

fig = px.bar(
    carencia_por_entidad,
    x='entidad',
    y='carencia_servicios',
    width=1200,
    height=600,
    title='Proporción de hogares con al menos una carencia básica por entidad',
    labels={'carencia_servicios': 'Proporción de hogares', 'entidad': 'Entidad federativa'},
    text=carencia_por_entidad['carencia_servicios'].round(2)
)

fig.update_layout(
    xaxis_tickangle=-45,
    yaxis_tickformat='.0%',
    template='plotly_dark'
)

fig.show()
```

###Proporción de hogares por tipo de piso

```
# Asegurarse de que todo esté como string limpia y consistente
caract_hogares['material_pisos'] = caract_hogares['material_pisos'].astype(str).str.strip()

# Diccionario de ejemplo para mapear códigos a etiquetas
mapa_pisos = {
    '1': 'Tierra',
    '2': 'Cemento',
    '3': 'Madera',
    '4': 'Loseta',
    '5': 'Otro'
}

caract_hogares['material_pisos'] = caract_hogares['material_pisos'].map(mapa_pisos).fillna('No especificado')
```

```
# Agrupar por entidad y tipo de material de pisos
freq_pisos = caract_hogares.groupby(['entidad', 'material_pisos']).size().reset_index(name='conteo')

# Obtener total de hogares por entidad
totales = freq_pisos.groupby('entidad')['conteo'].transform('sum')

# Agregar columna de porcentaje
freq_pisos['porcentaje'] = freq_pisos['conteo'] / totales

fig = px.bar(
    freq_pisos,
    x='entidad',
    y='porcentaje',
    width=1200,
    height=600,
    color='material_pisos',
    title='Distribución del tipo de material de pisos por entidad',
    labels={
        'porcentaje': 'Proporción de hogares',
        'entidad': 'Entidad federativa',
        'material_pisos': 'Material del piso'
    },
    text_auto='.1%',
    barmode='stack'
)

fig.update_layout(
    xaxis_tickangle=-45,
    yaxis_tickformat='.0%',
    template='plotly_dark'
)
fig.show()
```

###Características de las viviendas (por entidad)

```
# Diccionarios de mapeo
map_tipo_vivienda = {
    1: "Casa única",
    2: "Casa compartida",
    3: "Casa dúplex",
    4: "Departamento",
    5: "Vecindad",
    6: "Cuarto en azotea",
    7: "Local no habitable"
}

map_material_paredes = {
    1: "Desecho",
    2: "Lámina cartón",
    3: "Lámina metálica/asbesto",
    4: "Carrizo/palma",
    5: "Embarro/bajareque",
    6: "Madera",
    7: "Adobe",
    8: "Tabique/ladrillo/concreto"
}

map_material_techos = {
    1: "Desecho",
    2: "Lámina cartón",
    3: "Lámina metálica",
    4: "Asbesto",
    5: "Fibrocemento",
    6: "Palma/paja",
    7: "Madera",
    8: "Terrado",
    9: "Teja",
    10: "Losa concreto"
}

map_material_pisos = {
    1: "Tierra",
    2: "Cemento/firme",
    3: "Madera/mosaico/otro"
}

# Aplicar mapeo
caract_hogares['tipo_vivienda_txt'] = caract_hogares['tipo_vivienda'].map(map_tipo_vivienda)
caract_hogares['paredes_txt'] = caract_hogares['material_paredes'].map(map_material_paredes)
caract_hogares['techos_txt'] = caract_hogares['material_techos'].map(map_material_techos)
caract_hogares['pisos_txt'] = caract_hogares['material_pisos'].map(map_material_pisos)
```

```
# Distribución de tipo de vivienda por entidad

material_por_entidad = (
    caract_hogares
    .groupby(['entidad', 'paredes_txt'])
    .size()
    .reset_index(name='conteo')
)

# Calcular porcentaje por entidad
material_por_entidad['porcentaje'] = (
    material_por_entidad
    .groupby('entidad')['conteo']
    .transform(lambda x: (x / x.sum()) * 100)
)
```

```
fig = px.bar(
    material_por_entidad,
    x="entidad",
    y="porcentaje",
    width=1200,
    height=600,
    color="paredes_txt",
    title="Distribución de tipo de vivienda por entidad",
    labels={"porcentaje": "Porcentaje", "entidad": "Entidad federativa", "paredes_txt": "Tipo de material"},
    text_auto=".2s"
)

fig.update_layout(
    barmode="stack",
    xaxis_tickangle=-45,
    legend_title_text="Material de vivienda",
    template='plotly_dark'
)

fig.show()
```
